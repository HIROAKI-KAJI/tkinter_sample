## widget の表示

　tkinter のテキストやボタンプルダウンメニューなどをWidgetといい、親のフレームなどに紐付けることで表示することが出来る。
 表示の仕方にはgrid(),pack(),place()の３つある。

### grid()

 親となるフレームを表のように区切りその何行何列目に表示するかを決めて配置する方法

#### 例　sample というテキストをrootのフレームの0行０列目に配置

```python
import tkinter as tk

root = tk.Tk() 
tk.Label(root,text="sample").grid(row = 0,column =0)
root.mainloop()
```

#### sticky 表のセル中のどちら側に張り付くのかを決める

sticky = tk.N　のように記述<br>

方角で指定し、tk.N,tk,W,tk.E,tk.Sが指定できる。<br>

左上や、中心などを設定するときは、tk.N + tk.E　や　tk.N+tk.E+tk.W+tk.Sの様に指定する。

```python
import tkinter as tk

root = tk.Tk() 
tk.Label(root,text="Norse+West").grid(row = 0,column =0,sticky=tk.N+tk.W)
tk.Label(root,text="West").grid(      row = 1,column =0,sticky=tk.W)
tk.Label(root,text="Souse+West").grid(row = 2,column =0,sticky=tk.S+tk.W)

tk.Label(root,text="Norse").grid(     row = 0,column =1,sticky=tk.N)
tk.Label(root,text="center").grid(    row = 1,column =1,sticky=tk.N+tk.W+tk.E+tk.S)
tk.Label(root,text="Souse").grid(     row = 2,column =1,sticky=tk.S)

tk.Label(root,text="Norse+East").grid(row = 0,column =2,sticky=tk.N+tk.E)
tk.Label(root,text="East").grid(      row = 1,column =2,sticky=tk.E)
tk.Label(root,text="Souse+East").grid(row = 2,column =2,sticky=tk.S+tk.E)
root.mainloop()
```

#### rowspan,columnspan

　何列何行の幅を持たせるのかを設定できる。

```python
import tkinter as tk

root = tk.Tk() 
tk.Label(root,text="sample").grid(row = 0,column =0,rowspan = 4)

for n in range(4):
    tk.Label(root,text=f"{n}").grid(row = n,column =1)
root.mainloop()
```

### pack

　フレームの右や上などに詰めていくように配置する方法<br>
 (後から配置を変更するとき大変になる傾向がある)
 

```python
import tkinter as tk

root = tk.Tk() 
tk.Label(root,text="sample").pack()

root.mainloop()
```

#### pack side

どちら側から詰めていくのかを決めることができる

```python
import tkinter as tk

root = tk.Tk() 
tk.Label(root,text="top").pack(side = tk.TOP)
tk.Label(root,text="right").pack(side = tk.RIGHT)
tk.Label(root,text="left").pack(side = tk.LEFT)
tk.Label(root,text="bottom").pack(side = tk.BOTTOM)
root.mainloop()
```

配置された順番に埋めていくので、今回の場合は、以下のようにフレームを埋める。<br>


#### pack anchor

 配置可能な範囲でどちら側に寄るのかを設定<br>
 
 sticky と同じ引数を持たせることができる 

```python
import tkinter as tk

root = tk.Tk() 
tk.Label(root,text="sample").pack(anchor= tk.N)
tk.Label(root,text="sample").pack(anchor= tk.W)
tk.Label(root,text="sample").pack(anchor= tk.E)
tk.Label(root,text="sample").pack(anchor= tk.N)
root.mainloop()
```

### place 座標を決めて配置

 place は滅多に使用しない.. がキャンバスの上に配置したいときなどに使わないこともない。

```python
import tkinter as tk

root = tk.Tk() 
tk.Label(root,text="place").place(x = 30, y = 40)

root.mainloop()
```

## 各種widget

### label

```python
import tkinter as tk

root = tk.Tk() 

textv = tk.StringVar()
textv.set("可変テキスト")# Entryのところで説明

tk.Label(root,text="label").grid(row = 0,column =0)
tk.Label(root,textvariable=textv).grid(row = 1,column =0)
root.mainloop()
```

### button

command =  でボタンを押したとき実行するコマンドを設定できる。

```python
import tkinter as tk

def printa():
    print("A")

root = tk.Tk() 
tk.Button(root,text="button").grid(row = 0,column =0)
tk.Button(root,text="print\A\"",command = printa).grid(row = 1,column =0)
root.mainloop()
```

### Entry

```python
import tkinter as tk

root = tk.Tk()

text = tk.StringVar()
#可変文字列StringVar()　このほかにIntVarなどがある。値を取り出すときは、
# .get() で取り出せる。また、初期値などを設定するときは .set("text")などで
#決めることができる。

#注意点、このStringVarなどは、Tk()に紐づけられる。よって、root =Tk()のように、
# Tk()オブジェクトを呼び出した後でないと設定できない。

tk.Entry(root,textvariable=text).grid(row = 0,column =0)

root.mainloop()
```
